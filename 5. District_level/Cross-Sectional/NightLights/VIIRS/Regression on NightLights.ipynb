```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
%matplotlib inline
```

```python
features = pd.read_csv('VIIRS_2012_SumNightLights_100m.csv')

features.columns = ['census_code', 'sum_nightlights']
```

```python
ground_truth = pd.read_csv(r"C:\Users\AJain7\OneDrive - Stryker\Personal\Projects\Satellite Project\5. District_level\District_Master.csv")
cols = ['CHH_RUD','CHH_INT','CHH_ADV','district code']
ground_truth = ground_truth[cols]
```

```python
ground_truth.columns
```

```python
ground_truth.columns = ['CHH_RUD', 'CHH_INT', 'CHH_ADV', 'census_code']
```

```python
print(features.shape)
print(ground_truth.shape)
data = ground_truth.merge(features, on='census_code', how='left')
data.dropna(inplace=True)
print(data.shape)
```

```python
indicator = 'CHH_ADV'
```

```python
x = data['sum_nightlights'].values
y = data[indicator].values
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

## Model

```python
regression_model = LinearRegression()
regression_model.fit(x, y)
y_predicted = regression_model.predict(x)
rmse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)

print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
print('Total data points: ', data.shape[0])
```

```python
# data points
plt.figure(figsize=(10,10))
plt.scatter(x, y, s=10)
# predicted values
plt.plot(x, y_predicted, color='r')
plt.title(indicator+':  r2_score: '+ str(r2))
plt.show()
```
