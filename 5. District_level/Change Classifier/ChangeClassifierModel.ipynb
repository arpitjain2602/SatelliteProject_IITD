```python
folder = 'input_files/'
```

```python
data_dict = {}
for files in os.listdir(folder):
    file_name = os.path.join(folder,files)
    
    indicator = files.split('.')[0]
    
    if (indicator == 'ASSET'):
        print(indicator)
        columns = [indicator+'_2001', indicator+'_2003', indicator+'_2005', indicator+'_2007', indicator+'_2009', indicator+'_Actual_2001', 'Formal_EMP_2001', 'LIT_2001', indicator+'_change']
        df = pd.read_csv(file_name, usecols=columns)
        print(df.columns)
        data_dict['ASSET'] = df
    
    elif (indicator == 'BF'):
        print(indicator)
        columns = [indicator+'_2001', indicator+'_2003', indicator+'_2005', indicator+'_2007', indicator+'_2009', indicator+'_Actual_2001', 'Formal_EMP_2001', 'LIT_2001', indicator+'_change']
        df = pd.read_csv(file_name, usecols=columns)
        print(df.columns)
        data_dict['BF'] = df
    
    elif (indicator == 'EMP'):
        print(indicator)
        columns = [indicator+'_2001', indicator+'_2003', indicator+'_2005', indicator+'_2007', indicator+'_2009', indicator+'_Actual_2001',indicator+'_change']
        df = pd.read_csv(file_name, usecols=columns)
        print(df.columns)
        data_dict['EMP'] = df
        
    elif (indicator == 'FC'):
        print(indicator)
        columns = [indicator+'_2001', indicator+'_2003', indicator+'_2005', indicator+'_2007', indicator+'_2009', indicator+'_Actual_2001', 'Formal_EMP_2001',indicator+'_change']
        df = pd.read_csv(file_name, usecols=columns)
        print(df.columns)
        data_dict['FC'] = df
        
    elif (indicator == 'MSL'):
        print(indicator)
        columns = [indicator+'_2001', indicator+'_2003', indicator+'_2005', indicator+'_2007', indicator+'_2009', indicator+'_Actual_2001', indicator+'_change']
        df = pd.read_csv(file_name, usecols=columns)
        print(df.columns)
        data_dict['MSL'] = df
        
    elif (indicator == 'MSW'):
        print(indicator)
        columns = [indicator+'_2001', indicator+'_2003', indicator+'_2005', indicator+'_2007', indicator+'_2009', indicator+'_Actual_2001', indicator+'_change']
        df = pd.read_csv(file_name, usecols=columns)
        print(df.columns)
        data_dict['MSW'] = df
```
